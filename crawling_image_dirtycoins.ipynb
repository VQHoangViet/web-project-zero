```python
# Get link to full of image and information about product
wd.get('https://dirtycoins.vn/sale-2')
s = wd.find_elements_by_xpath('//div[@class = "product-img img-full "]//a')
in4_links = []

for i in range(1,6):
  wd.get('https://dirtycoins.vn/sale-2?q=collections:2756598&page='+ str(i) +'&view=grid')
  s = wd.find_elements_by_xpath('//div[@class = "product-img img-full "]//a')
  for i in s:
    in4_links.append(i.get_attribute('href'))
    print(in4_links[-1])
```

```python
# Get infors and images
title_txt = []
price_txt = []
image_links_txt = []
des_product_txt = []
product_thumb_slider = []

for i in in4_links:
  wd.get(i)
  image_links_txt.append(wd.find_element_by_xpath('//a[@class = "fancybox-thumb"]//img').get_attribute('src'))
  price_txt.append(wd.find_element_by_xpath('//span[@class ="price product-price"]').text)
  title_txt.append(wd.find_element_by_xpath('//h1[@class ="title-product"]').text)
  des_product_txt.append(wd.find_element_by_xpath('//div[@class="product-detail-content"]').text)
  
  temp_product_thumb_slider=[]
  els = wd.find_elements_by_xpath('//li[@class="product-thumb "]//img')
  for i in els:
    temp_product_thumb_slider.append(i.get_attribute('src'))
  product_thumb_slider.append(temp_product_thumb_slider)

  print(str(title_txt[-1]) + ': ' + 
        str(price_txt[-1]) + "\n" +
        str(des_product_txt[-1]) + '\n' + "Links_img:" +
        str(image_links_txt[-1]) + "\n" +
        str(product_thumb_slider[-1]) + "\n" +
        "-"*30)

```

```python
# to DataFrame
album_ = pd.DataFrame({
    'Title': title_txt,
    'Links': image_links_txt,
    'Price': price_txt,
    'Product thumb': list(product_thumb_slider)

})
```

```python
price_str = []
for i in album_.Price:
  x = i.split('₫')
  price_str.append(x[0])

price_num = []
for i in price_str:
  try:
    price_num.append(float(i)*1000)
  except:
    price_num.append(str(i))

album_.Price = price_num
```

```python
album_lien_he = album_[album_['Price'] == 'Liên hệ']
album_= album_[album_['Price'] != 'Liên hệ']
```

```python
def downloadImage(list_url, filepath, count=0):
  for url in list_url:
    dir = filepath + '/product' + str(count)
    os.mkdir(dir)
    file_name = wget.download(url, out=dir)
    print('[' + str(count) + '] '+ 'Download succesfull: ' + file_name)
    count +=1
```

```python
downloadImage(list_url=album_['Links'], filepath='/content/img')
```

```python
count=0
for links in album_['Product thumb']: 
  dir ='/content/img/product' + str(count)
  for link in links:
    file_name = wget.download(link, out=dir)
    print('Download Success: ' + file_name)
  count +=1
```
